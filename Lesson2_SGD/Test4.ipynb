```python
import numpy as np
import torch
import pandas as pd
from torch import tensor
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import requests
r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/inflation?period=m&json')
#r.json()
```

```python
with open('data_infl.json', 'wb') as fd:
    for chunk in r.iter_content(chunk_size=128):
        fd.write(chunk)
```

```python
df = pd.read_json("data_infl.json")
#df.to_csv()
```

```python
df.to_csv('data_infl.csv', encoding='utf-8', index=False)
```
