# Max close price

```python
def get_max_close(symbol):
    df = pd.read_csv("{}.csv".format(symbol))
    return df['Close'].max()
```

```python
def symbols():
    for symbol in ['AAPL', 'FB']:
        print("Max close")
        print(symbol, get_max_close(symbol))
```

```python
def get_data(symbols, dates):
    df1 = pd.DataFrame(index=dates)
    for symbol in symbols:
      df_tmp = pd.read_csv("{}.csv".format(symbol), index_col="Date", parse_dates=True,
                         usecols=['Date', 'Close', 'Adj Close'], na_values=['nan'])
      df_tmp = df_tmp.rename(columns={'Close': symbol + ' ''Close', 'Adj Close':symbol  + ' ''Adj Close'})
      df1 = df1.join(df_tmp)
    
    df1 = df1.dropna() #drops all na values
    return df1

def get_df():
    start_date='2019-01-01'
    end_date='2019-12-31'
    dates = pd.date_range(start_date, end_date)
    
    symbols = ['GOOG', 'FB']
    
    df = get_data(symbols, dates)
    
    return df.loc['2019-01-01':'2019-01-31', ['GOOG Close', 'FB Close']]

def plot_data(df, title="Stock prices"):
    ax = df.plot(title=title, fontsize=12)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
```

# Normalized plot

```python
def get_data(symbols, dates):
    df1 = pd.DataFrame(index=dates)
    for symbol in symbols:
      df_tmp = pd.read_csv("{}.csv".format(symbol), index_col="Date", parse_dates=True,
                         usecols=['Date', 'Close', 'Adj Close'], na_values=['nan'])
      df_tmp = df_tmp.rename(columns={'Close': symbol + ' ''Close', 'Adj Close':symbol  + ' ''Adj Close'})
      df1 = df1.join(df_tmp)
    
    df1 = df1.dropna() #drops all na values
    return df1

def get_df():
    start_date='2019-01-01'
    end_date='2019-12-31'
    dates = pd.date_range(start_date, end_date)
    
    symbols = ['GOOG', 'FB']
    
    df = get_data(symbols, dates)
    
    return df.loc['2019-01-01':'2019-01-31', ['GOOG Close', 'FB Close']]

def normalize_data(df):
    return df/ df.iloc[0] #dividing all prices by the first one

def plot_data(df, title="Normalized stock prices"):
    ax = df.plot(title=title, fontsize=10)
    ax.set_xlabel("Date", fontsize=14)
    ax.set_ylabel("Price", fontsize=14)
    plt.show()
```

```python
get_df()
```

```python
normalize_data(get_df())
```

```python
plot_data(normalize_data(get_df()))
```
