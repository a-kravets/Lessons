# Slicing

```python
def slicing():
    df = pd.read_csv("AAPL.csv")
    print (df[10:21]) #rows from 10 to 20
```

```python
slicing()
```

# Max close price

```python
def get_max_close(symbol):
    df = pd.read_csv("{}.csv".format(symbol))
    return df['Close'].max()
```

```python
def symbols():
    for symbol in ['AAPL', 'FB']:
        print("Max close")
        print(symbol, get_max_close(symbol))
```

```python
symbols()
```
